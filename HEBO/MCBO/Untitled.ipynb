```python



def par_labels(arr_layers):
    products = [arr_layers[i] * arr_layers[i + 1] for i in range(len(arr_layers) - 1)]
    W_res_layers = [f'W{i}' for i, val in enumerate(products) for _ in range(val)]
    b_res_layers = [f'b{i}' for i, val in enumerate(arr_layers[1:]) for _ in range(val)]
    W_sub = [W_res_layers[sum(products[:i]):sum(products[:i+1])] for i in range(len(products))]
    b_sub = [b_res_layers[sum(arr_layers[1:i]):sum(arr_layers[1:i+1])] for i in range(1,len(arr))]
    W_indices, b_indices = [], []
    for layer in range(1,len(arr_layers)):
        #print(arr[layer])
        W_ind = ([[ f'_{i}{j}' for i in range(max(arr_layers)) for j in range(arr_layers[layer])][:products[k]] for k in range(len(products)) ])[layer-1]
        b_ind =  ( [ f'_{i}' for i in range(5)  ][:arr_layers[layer]] )
        W_indices.append(W_ind)
        b_indices.append(b_ind)
    W_elements = [[elem + index for elem, index in zip(sub, indices)] for sub, indices in zip(W_sub, W_indices)]
    b_elements = [[elem + index for elem, index in zip(sub, indices)] for sub, indices in zip(b_sub, b_indices)]
    return W_elements, b_elements


def label_parameters(layer_sizes):
    labels = {}
    for i in range(len(layer_sizes) - 1):
        labels[f"W{i}"] = f"{i}"
        labels[f"b{i}"] = f"{i}"
    return labels

import numpy as np

def initialize_parameters(layer_sizes):
    parameters = {}
    for i in range(0, len(layer_sizes)):
        parameters[f"W{i}"] = np.random.randn(layer_sizes[i], layer_sizes[i-1]) * 0.01
        parameters[f"b{i}"] = np.zeros((layer_sizes[i], 1))
    return parameters.keys()

def parameters_dict(params_array):
    """
    Create a dictionary of parameters from an array containing weights and biases.

    Parameters:
    params_array -- array containing weights and biases in the order [W0, b0, W1, b1, ..., Wn, bn]

    Returns:
    parameters -- dictionary containing the parameters 'W0', 'b0', 'W1', 'b1', ..., 'Wn', 'bn'
    """
    num_layers = len(params_array) // 2
    parameters = {}
    for i in range(num_layers):
        parameters[f'W{i}'] = params_array[2*i]
        parameters[f'b{i}'] = params_array[2*i + 1]
    return parameters



def forward_prop(X, parameters):
    """
    Perform forward propagation through each layer of a neural network.

    Parameters:
    parameters -- dictionary containing the parameters 'W0', 'b0', 'W1', 'b1', ..., 'Wn', 'bn'
    X -- input data (numpy array) of shape (input_size, m)

    Returns:
    A -- output of the last layer
    caches -- list of tuples containing the linear and activation values for each layer
    """
    caches = []
    A = X.T
    
    num_layers = len(parameters) // 2
    for i in range(num_layers):
        W = parameters[f'W{i}']
        b = parameters[f'b{i}']
        Z = np.dot(W, A) + b
        A = relu(Z) if i < num_layers - 1 else sigmoid(Z)
        caches.append((Z, A))

    return A.T, caches


# Activation functions
def relu(Z):
    return np.maximum(0, Z)

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

# Example usage:
#layer_sizes = [4, 5, 3, 1]

#parameter_labels = label_parameters(layer_sizes)  # Using previously defined label_parameters function
#parameters =np.array(list( parameter_labels.keys()))
#print((parameters))



def softmax(Z):
    return np.exp(Z) / np.sum(np.exp(Z))

# Loss function
def binary_cross_entropy(y_true, y_pred):
    cost = np.mean(np.abs(y_true - y_pred))  # Example: Mean absolute error
    return cost

def generate_sizes(layer_sizes):
    sizes = []
    start_idx = 0
    for size in layer_sizes:
        end_idx = start_idx + size
        sizes.append(end_idx - start_idx)
        start_idx = end_idx
    return sizes


def black_box_function(X_data, parameters):
    result = forward_prop(X_data, parameters)[0]
    y_train_encoded = np.array([y_train]) # np.eye(2)[y_train]
    return binary_cross_entropy(y_train_encoded, result)

def blocks(layer_sizes):
    blocks_info = []
    start_idx = 0
    for i in range(len(layer_sizes) - 1):
        # Weight block
        weight_block_size = layer_sizes[i] * layer_sizes[i+1]
        weight_block = (start_idx, start_idx + weight_block_size)
        reshape_weight = (layer_sizes[i+1], layer_sizes[i])
        
        # Bias block
        bias_block = (weight_block[1], weight_block[1] + layer_sizes[i+1])
        reshape_bias = (layer_sizes[i+1], 1)
        
        blocks_info.extend([weight_block, reshape_weight, bias_block, reshape_bias])
        
        start_idx = bias_block[1]  # Update start index for next iteration
        
    slicing, reshaping = blocks_info[::2], blocks_info[1::2] 
    #print(slicing, reshaping, 'slicing reshaping')
    return slicing, reshaping




class CustomTask(TaskBase):
    def __init__(self, pars, layer_sizes):
        self.layer_sizes = layer_sizes
        self.pars = [np.array([]) for _ in range(len(layer_sizes) - 1)] #pars 
        

    @property
    def name(self) -> str:
        return 'Custom Task'
    
    def evaluate(self, x: np.ndarray) -> Tuple[pd.DataFrame ,np.ndarray]:
        slices, reshape_dims = blocks(self.layer_sizes)[0], blocks(self.layer_sizes)[1]
        y = np.zeros((len(x), 1))
        for ind in range(len(x)):
            x_ind = x.iloc[ind].to_numpy()  # Convert Series to NumPy array
            sliced_arrays = [x_ind[start:end] for start, end in slices]
            reshaped_arrays = tuple([sliced_array.reshape(shape) 
                               for sliced_array, shape in zip(sliced_arrays, reshape_dims)])
            y[ind] = black_box_function(X_train, parameters_dict(reshaped_arrays))
        
        return x, y, parameters_dict(reshaped_arrays)
    
    def get_search_space_params(self) -> List[Dict[str, Any]]:
        params = []
        start_idx = 0
        for layer, size in enumerate(self.layer_sizes[:-1]):
            for i in range(size):
                for j in range(self.layer_sizes[layer + 1]):
                    params.append({'name': f'W{layer}_{i}{j}', 'type': 'num', 'lb': -1, 'ub': 1})
            for i in range(self.layer_sizes[layer + 1]):
                params.append({'name': f'b{layer}_{i}', 'type': 'num', 'lb': -1, 'ub': 1})
        return params
        

    def get_parameter_names(self) -> List[str]:
        params = []
        for layer, size in enumerate(self.layer_sizes[:-1]):
            for i in range(size):
                for j in range(self.layer_sizes[layer + 1]):
                    params.append(f'W{layer}_{i}{j}')
            for i in range(self.layer_sizes[layer + 1]):
                params.append(f'b{layer}_{i}')
        return params
        
# Creating task and optimization
task = CustomTask(parameters, layer_sizes)
searchspace = task.get_search_space()

epochs = 100
bo_builder = BoBuilder(model_id='gp_rd', acq_opt_id='is', acq_func_id='ei', tr_id='basic')
opt = bo_builder.build_bo(search_space=task.get_search_space(), n_init=200)


# Optimization loop
budget_eval = epochs
# Inside the optimization loop
weights_list = []
acc_list = []

for _ in range(budget_eval):
    x_next = opt.suggest()
    
    x_next_reshaped = np.array(x_next).reshape(1, -1)  # Reshape x_next to have shape (1, 25)
    #print(task.get_parameter_names(), 'x nextttttttttt')
    x_next_df = pd.DataFrame(x_next_reshaped, columns=task.get_parameter_names())
    y_next = task.evaluate(x_next_df)[1]
    weights_list.append(task.evaluate(x_next_df)[2])
    acc_list.append(y_next)
    opt.observe(x_next, y_next)
    print(y_next)
   


# Printing best result
print('Best parameters:', opt.best_x)
print('Best loss:', opt.best_y)

# Plotting version
fig, ax = plt.subplots(figsize=(7, 5))
y = opt.data_buffer.y.numpy()
regret_y = np.min(np.cumsum(y, axis=1), axis=0)
#plt.plot(np.arange(1,  budget_eval ), regret_y)
#plt.show()
```

```python
def binarize_outcome(res_arr):
    return np.where(res_arr<.5,0,1)

def calculate_accuracy( targets, predictions):
    """
    Calculate the accuracy between two arrays.

    Parameters:
    predictions -- numpy array containing predicted values
    targets -- numpy array containing true target values

    Returns:
    accuracy -- the accuracy between the predictions and targets
    """
    # Ensure predictions and targets have the same shape
    
    assert predictions.shape == targets.shape, "Shape mismatch between predictions and targets"

    # Count the number of correct predictions
    num_correct = np.sum(predictions == targets)

    # Calculate accuracy
    accuracy = num_correct / predictions.size

    return accuracy

def forward_prop(X, parameters):
    """
    Perform forward propagation through each layer of a neural network.

    Parameters:
    parameters -- dictionary containing the parameters 'W0', 'b0', 'W1', 'b1', ..., 'Wn', 'bn'
    X -- input data (numpy array) of shape (input_size, m)

    Returns:
    A -- output of the last layer
    caches -- list of tuples containing the linear and activation values for each layer
    """
    caches = []
    A = X.T
    
    num_layers = len(parameters) // 2
    for i in range(num_layers):
        W = parameters[f'W{i}']
        b = parameters[f'b{i}']
        Z = np.dot(W, A) + b
        A = relu(Z) if i < num_layers - 1 else sigmoid(Z)
        caches.append((Z, A))

    return A


binary_list = []
acc_list= []
for pars in weights_list:
    output = forward_prop(X_train, pars)
    #print((output.shape))
    binary_result = binarize_outcome(output)
    binary_list.append(binary_result)
    accuracy =calculate_accuracy(np.array(y_train), np.squeeze(binary_result))
    acc_list.append(accuracy)

print(max(acc_list))
print(acc_list.index(max(acc_list)))
```
